```python
# keep business_id, name, stars, review_count - use different variable to perserve the original df in case of future use
bdf2 = bdf[['business_id','name', 'stars', 'review_count']]
bdf2.head(10)
```

```python
# we want to merge using buisness_id, using pd.merge() on 'inner' and have one Data Frame
# mdf = merged data frames using bdf2 and dfg
# USE THIS TO LATER COMPARE RESULTS - just pd.concat() predicted onto this 

mdf =  pd.merge(bdf2, dfg, how = 'inner', on = 'business_id')
mdf
```

```python
# Now that we have that df we can choose the features that we want in this case
# I want to use stars as output, and review_count and text as input
# rdf = refined data frame -  contains text, review_count, and stars
rdf = mdf[['text', 'review_count', 'stars']]
rdf
```

```python
# Normalize review_count
# Encode a numeric column as zscores
def encode_numeric_zscore(df, name, mean=None, sd=None):
    if mean is None:
        mean = df[name].mean()

    if sd is None:
        sd = df[name].std()

    df[name] = (df[name] - mean) / sd

encode_numeric_zscore(rdf, 'review_count', mean=None, sd=None)
```

```python
# The review_count column is now normalized - to reduce the error
rdf
```

```python
# Know I want to use sklearns tfidfVectorizer to get the most uselful and most frequent words
import sklearn.feature_extraction.text as sk_text

vectorizer = sk_text.TfidfVectorizer(stop_words='english',
                             max_features = 8000,
                             min_df= 2, 
                             #max_df=5
                             )

matrix = vectorizer.fit_transform(rdf['text'])

print(type(matrix))         
text_to_tfidf = matrix.toarray()
print(text_to_tfidf)
```

```python
print(vectorizer.get_feature_names())
```

```python
# check to see if these words match the review words, which they do for example tdidf_date at index 3 is a photo business
# we can use inverse_transform to idetify the tf_idf results back to original and see values turned back to words
vectorizer.inverse_transform(text_to_tfidf[3])
```

```python
# We want to turn review_count into a numpy array

review_count = np.array(rdf['review_count'])
print(type(review_count))
print(review_count)
```

```python
# Then we want to concat the arrays together
# This will be our x value
# np.column_stack(([1,2,3],[4,5,6])) ex. will look like this ...
# array([[1, 4],
#       [2, 5],
#       [3, 6]])
# So that we can pass all of it into test train split as a single variable
x_values = np.column_stack((text_to_tfidf, review_count))
type(x_values)
```

```python
# Remember that our target output is the column stars from rdf - which is what we want to use as out y
rdf['stars']
```

```python
from sklearn.model_selection import train_test_split


# Split into train/test - we have to import train_test_split, the parameters will be x_values: which will be split into x_train, x_train
# rdf['stars'] will be the second parameter: which will be split into y_train, y_test
# test will be 25%, train will be 75% of our data
# we want rdf['stars'].values (gives array) which will only get the rating which is what we want.

x_train, x_test, y_train, y_test = train_test_split(x_values, rdf['stars'].values, test_size=0.25, random_state=42)
```

```python
import tensorflow as tf
from sklearn import metrics
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint

# Now that our data is inspected we can continue, the next step would be to set up our checkpoint
# to save the best model. Remember: define ModelCheckpoint outside the loop
checkpointer = ModelCheckpoint(filepath="/Users/robertocampos/Desktop/Main_Folder/Dataset/yelp_dataset/dnn/best_weights.hdf5", verbose=0, save_best_only=True)

# for i in range(5):
#     print(i)

#     # Build network
#     model = Sequential()
#     model.add(Dense(30, input_dim=x_train.shape[1], activation='relu'))
#     model.add(Dense(15, activation='relu'))
#     model.add(Dense(10, activation='relu'))
#     model.add(Dense(2,  activation='sigmoid'))
#     model.add(Dense(1))
#     model.compile(loss='mean_squared_error', optimizer='sgd')

#     monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=5, verbose=1, mode='auto')

#     model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[monitor,checkpointer],verbose=2,epochs=110)
    
#     # Root mean square error - On average how for is predicted y from actual y
#     y_pred = model.predict(x_test)
#     score = np.sqrt(metrics.mean_squared_error(y_pred,y_test))
#     print("Score (RMSE): {}".format(score))

# print('Training finished...Loading the best model')  
# print()
for i in range(5):
    print(i)

     # Build network
    model = Sequential()
    model.add(Dense(30, input_dim=x_train.shape[1], activation='relu'))
    model.add(Dense(15, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(2,  activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=5, verbose=1, mode='auto')

    model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[monitor,checkpointer],verbose=2,epochs=110)
    
     # Root mean square error - On average how for is predicted y from actual y
    y_pred = model.predict(x_test)
    score = np.sqrt(metrics.mean_squared_error(y_pred,y_test))
    print("Score (RMSE): {}".format(score))

print('Training finished...Loading the best model')  
print()

# load weights from best model
model.load_weights('/Users/robertocampos/Desktop/Main_Folder/Dataset/yelp_dataset/dnn/best_weights.hdf5') 
```

```python
# Regression chart 
def chart_regression(pred,y,sort=True):
    t = pd.DataFrame({'pred' : pred, 'y' : y.flatten()})
    if sort:
        t.sort_values(by=['y'],inplace=True)
    b = plt.plot(t['pred'].tolist(),label='prediction')
    a = plt.plot(t['y'].tolist(),label='expected')
    plt.ylabel('output')
    plt.legend()
    plt.show()

chart_regression(y_pred.flatten(), y_test, sort = True)
```
